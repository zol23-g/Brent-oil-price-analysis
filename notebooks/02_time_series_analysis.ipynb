### Reshape Each Dataset from Wide to Long Format

Each dataset has columns for every year. We need to:

Melt these columns into Year and Value columns.
Filter for specific years if needed.

```python
# Reshape GDP data
gdp_data = gdp_data.melt(id_vars=['Country Name', 'Country Code'], 
                         value_vars=[str(year) for year in range(1987, 2023)],
                         var_name='Year', value_name='GDP Growth Rate')

# Convert 'Year' to datetime format and set it as index
gdp_data['Year'] = pd.to_datetime(gdp_data['Year'], format='%Y')
gdp_data.set_index('Year', inplace=True)
print(gdp_data.head())
```

```python
# Reshape Inflation data
inflation_data = inflation_data.melt(id_vars=['Country Name', 'Country Code'], 
                                     value_vars=[str(year) for year in range(1987, 2023)],
                                     var_name='Year', value_name='Inflation Rate')
inflation_data['Year'] = pd.to_datetime(inflation_data['Year'], format='%Y')
inflation_data.set_index('Year', inplace=True)
print(inflation_data.head())


```

```python
# Reshape Exchange Rate data
exchange_rate_data = exchange_rate_data.melt(id_vars=['Country Name', 'Country Code'], 
                                             value_vars=[str(year) for year in range(1987, 2023)],
                                             var_name='Year', value_name='Exchange Rate')
exchange_rate_data['Year'] = pd.to_datetime(exchange_rate_data['Year'], format='%Y')
exchange_rate_data.set_index('Year', inplace=True)
print(exchange_rate_data.head())
```

### Merge Data with Brent Oil Prices
#### Forward-fill Economic Data to Daily Frequency

```python
# Forward-fill the annual economic data to match the daily frequency of Brent oil prices
gdp_data_daily = gdp_data.resample('D').ffill()
inflation_data_daily = inflation_data.resample('D').ffill()
exchange_rate_data_daily = exchange_rate_data.resample('D').ffill()

# Merge daily Brent oil prices with forward-filled economic data
merged_data_daily = brent_data.join([gdp_data_daily['GDP Growth Rate'],
                                     inflation_data_daily['Inflation Rate'],
                                     exchange_rate_data_daily['Exchange Rate']],
                                    how='left')

# Display the merged data
print("Merged Data (Daily):")
print(merged_data_daily.head())


```

### Modeling the Data

```python
from statsmodels.tsa.api import VAR

# Selecting relevant columns for the VAR model
var_data = merged_data_daily[['Price', 'GDP Growth Rate', 'Inflation Rate', 'Exchange Rate']].dropna()

# Fit the VAR model
model = VAR(var_data)
var_result = model.fit(2)  # Using lag order 2 as an example
print(var_result.summary())

# Forecasting using VAR
forecast = var_result.forecast(var_data.values[-2:], steps=10)
forecast_df = pd.DataFrame(forecast, index=pd.date_range(start=var_data.index[-1], periods=10, freq='D'),
                           columns=var_data.columns)
print("VAR Forecast:")
print(forecast_df)
```

### Long Short-Term Memory (LSTM) Neural Network

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler

# Scale data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(merged_data_daily['Price'].values.reshape(-1,1))

# Prepare data for LSTM
X, y = [], []
for i in range(60, len(scaled_data)):
    X.append(scaled_data[i-60:i, 0])
    y.append(scaled_data[i, 0])

X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Build LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X, y, epochs=20, batch_size=32)

# Predictions
predictions = model.predict(X)
predicted_prices = scaler.inverse_transform(predictions)
```
